```python
from datetime import datetime
from revision_fechas import revision_fecha
import warnings

warnings.filterwarnings('ignore')

lista_revision = []

# fecha_inicio = datetime.strptime('19-06-2023', "%d-%m-%Y")
# fecha_fin =  datetime.strptime('31-10-2023', "%d-%m-%Y")

for sujeto in lista_sujetos:
    sujeto_principal = df[df['Participante']==sujeto]

    visit_dictionary = {'Screening Visit': '', 'D-1':'', 'D1':'', 'D2':'', 'D3':'', 'D4':'', 'D7':'', 'D14':'', 
    'D15':'', 'D16':'', 'D17':'', 'D18':'', 'D21':'', 'D28':'', 'D29':'', 'D30':'',
    'D31':'', 'D32':'', 'D35':'', 'D42':'', 'D63':'', 'D90':'', 'D105':''}

    for visita in sujeto_principal.Visit.unique():
     
        pru_1 = sujeto_principal[sujeto_principal['Visit']==visita]
        pru = pru_1
        pru = pru[['Campo', 'Valor']].T
        new_columns = pru.iloc[0]
        pru = pru[1:].set_axis(new_columns, axis=1)
        pru['Subject'] = sujeto
        pru['Visit'] = visita
        pru['status'] = pru_1['activityState'].unique()

        pru = pru.merge(df_informed, on=['Subject', 'Visit'], how='left')
        

        for index, row in pru.iterrows():
            status = row['status']
            try:
                was_performed = row['Was the visit performed?']
            except:
                pass
            provide_reason = row['Provide the reason']
            visit_date = row['Visit Date']
            subject = row['Subject']
            visit = row['Visit']
            informed_consent_date = row['Valor']

            
            
            if status == 'INCOMPLETE' or status == 'PENDING_QUERIES':
                pass
            else:
                
                # Primera  revision general de formato de fecha ->GE0020
                f = revision_fecha(visit_date)
                if f == None:
                    pass
                else:
                    error = [subject, visit, f , visit_date, 'GE0020']
                    lista_revision.append(error)         

                # Agregamos la fecha de visit date al diccionario, para poder ser consultado posteriormente
                visit_dictionary[visit] = visit_date
                
                try:
                    # Verificamos si la fecha de la visita es mayor o igual al inofrm consent date -> VS0050
                    if datetime.strptime(str(visit_date), '%d-%b-%Y') >= datetime.strptime(str(informed_consent_date), '%d-%b-%Y'):
                        pass
                    else:
                        error = [subject, visit, 'Visit date must be equal or greater than the inform consent date', visit_date, 'VS0050']
                        lista_revision.append(error)
                except:
                    pass

    
    # Codigo para verificar si las fechas de las visitas, son mayores consecutivamente -> VS0020
    lista_valores_diccionario_visitas = list(visit_dictionary.values())
    lista_keys_diccionario_visitas = list(visit_dictionary.keys())

    for i in range(len(lista_valores_diccionario_visitas)):

        if  lista_valores_diccionario_visitas[i] == 0:
            pass
        elif lista_valores_diccionario_visitas[i] == '' :
            break

        elif lista_valores_diccionario_visitas[i] == 'D-1' :
                if datetime.strptime(lista_valores_diccionario_visitas[i], '%d-%b-%Y')  < datetime.strptime(lista_valores_diccionario_visitas[i-1], '%d-%b-%Y'):
                    error = [sujeto, lista_keys_diccionario_visitas[i], 'Visit date must be greater than the previous visit date' , lista_valores_diccionario_visitas[i], 'VS0020']
                    lista_revision.append(error)
                else:
                    pass
            

        else:
            try:
                if datetime.strptime(lista_valores_diccionario_visitas[i], '%d-%b-%Y')  <= datetime.strptime(lista_valores_diccionario_visitas[i-1], '%d-%b-%Y'):
                    error = [sujeto, lista_keys_diccionario_visitas[i], 'Visit date must be greater than the previous visit date' , lista_valores_diccionario_visitas[i], 'VS0020']
                    lista_revision.append(error)
                else:
                    pass
            except:
                pass



```
